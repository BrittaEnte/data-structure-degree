```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

    def __repr__(self):
        return str(self.value)


class LinkedList:
    def __init__(self):
        self.head = None

    def __str__(self):
        cur_head = self.head
        out_string = ""
        while cur_head:
            out_string += str(cur_head.value) + " -> "
            cur_head = cur_head.next
        return out_string


    def append(self, value):

        if self.head is None:
            self.head = Node(value)
            return

        node = self.head
        while node.next:
            node = node.next

        node.next = Node(value)

    def size(self):
        size = 0
        node = self.head
        while node:
            size += 1
            node = node.next

        return size

def union(llist_1, llist_2):
    # Your Solution Here
    '''
    we will check for each list if there are duplicates in it. 
    the duplicates will not added to our new list. 
    then we have one list which got the figures of the two lists. 
    '''
    
    values = set() # in a set there are no duplicates 
    node = llist_1.head # set node as head
    new_list = LinkedList() 
    
    while node:
        if node.value not in values:
            values.add(node.value)
            new_list.append(node)
         # if value is in the list, just go to the next node  
        node = node.next
    
    node = llist_2.head # set node as head
    while node:  
        if node.value not in values:
            values.add(node.value)
            new_list.append(node)
            
        node = node.next
    return new_list 

def intersection(llist_1, llist_2):
    # Your Solution Here
    ''''
    first we do the same as before. we put all value one time into the set
    then we we turn the game around. if the value of the second list, is
    already in the set we remove this from the set, but we add it to the list
    '''
    #new list 
    values = set()
    node = llist_1.head
    new_list = LinkedList()
    while node:
        if node.value not in values:
            values.add(node.value)
        node = node.next
    
    node = llist_2.head
    while node:
        if node.value in values:
            values.remove(node.value)
            new_list.append(node)
        node = node.next
            
    return new_list


# Test case 1

linked_list_1 = LinkedList()
linked_list_2 = LinkedList()

value_1 = [3,2,4,35,6,65,6,4,3,21]
value_2 = [6,32,4,9,6,1,11,21,1]

for i in value_1:
    linked_list_1.append(i)

for i in value_2:
    linked_list_2.append(i)

print (f"union {union(linked_list_1,linked_list_2)}")

print (f"intersection {intersection(linked_list_1,linked_list_2)}")

# Test case 2

linked_list_5 = LinkedList()
linked_list_6 = LinkedList()

value_1 = [3,2,4,35,6,65,6,4,3,23]
value_2 = [1,7,8,9,11,21,1]

for i in value_1:
    linked_list_5.append(i)

for i in value_2:
    linked_list_6.append(i)

print (f"union {union(linked_list_5,linked_list_6)}")

print (f"intersection {intersection(linked_list_5,linked_list_6)}")

# Test case 3

linked_list_3 = LinkedList()
linked_list_4 = LinkedList()

value_1 = []
value_2 = []

for i in value_1:
    linked_list_3.append(i)

for i in value_2:
    linked_list_4.append(i)

print (f"union {union(linked_list_3,linked_list_4)}")

print (f"intersection {intersection(linked_list_3,linked_list_4)}")

```
